**Assesing elements in a list**

```python
bicycles = ['car', 'bus', 'train', ' plane']
print(bicycles[0])
print(bicycles[1].title())
print(bicycles[3])
print(bicycles[3].lstrip().title())
print(bicycles)
```

**Index position start at 0, not 1**

For assessing list elements from last ; -1 is used for last, -2 is used for assessing second last element from list and so on.

```python
students = ['ram', 'shyam', 'hari', 'rajan', 'narayan']
print(students[-1]) #output:narayan
print(students[-2]) #output:rajan
print(students[-3]) #output:hari
```

**changing , adding and removing list elements**

Modifying elements in a list

```python
bikes = ['honda', 'yamaha', 'fz',]
print(bikes)
bikes[0] = 'duagti'
bikes[1] = 'bmw'
print(bikes)
#honda is replaced by dugati and yamaha is replaced by bmw.
```

Adding elements to a list.

```python
#Using append to add a element in list in last.
print("For adding elements to the end of the list we use append.")
bikes = ['honda', 'yamaha', 'fz']
bikes.append('dugati')
bikes.append('bmw')
print(bikes)
```

```python
#Adding elements in a  completely empty list i.e, motorcycle[]. 
motorcycles = []
motorcycles.append('fz')
motorcycles.append('honda')
motorcycles.append('dugati')
print(motorcycles)
```

```python
#you can add a new element at any position in your list by using the insert() method
cars = ['toyota', 'maruti', 'suzuki']
cars.insert(0, 'mercedes')
cars.insert(2, 'roalse royalce') #ignore the spelling
print(cars)
```

**Removing elements from a list**

If you know the position of the item you want to remove from a list, you can use the del statement.


```python
cars1 = ['toyota','mercedes', 'lambourgini']
del cars1[0]
del cars1[1]
print(cars1)
```

**Removing the item using the pop() method**

Sometimes you will want to use the value of an item after you remove it from a list.The pop() method removes the last item from a list,but it lets you work with that item after removing it.


```python
planes = ['airbus', 'boeing', 'lockheed martin']
print(planes)
popped_planes = planes.pop()
print(planes)
print(popped_planes)
```
