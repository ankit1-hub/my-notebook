**Removing an item using pop() method**

```python
friends = ['ram','shyam','hari','rajan']
a = friends.pop()
print(friends)
print(a)
```

we still have the acess to the element after removing it from the list.


**Popping items from any position in a list**

```python
students1 = ['rajan', 'hari', 'narayan','faran']
students1.pop(2)
print(students1)
```

**Removing an item by value**

sometimes we don't know the position .so, we can remove any value by its name using remove() method.

```python
freefire = ['peak', 'mill', 'factory', 'mars_electric']
dangerous = 'peak'
freefire.remove(dangerous)
print(freefire)
print(f"{dangerous.title()} is so dangerous.This is only for pro players.")

```
