```python
language = "python "
language1 = "c "
print(language.rstrip())
print(language1.rstrip())
#rstrip() is only used for removing right space not left space 
print(language)
language = language.rstrip()
print(language)
```

```python
language = " python "
print(language.lstrip())
print(language.strip())
```
